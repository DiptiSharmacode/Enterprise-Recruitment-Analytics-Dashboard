```python
import pandas as pd
import numpy as np

employees = pd.read_csv(
    "../datasets/dimensions/Dim_Employee.csv"
)

print(employees.shape)
```

```python
salary_lookup = {
    1: (300000, 600000),
    2: (600000, 1200000),
    3: (1200000, 2500000),
    4: (2500000, 5000000)
}
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

months = pd.date_range(
    start="2023-01-01",
    end="2025-12-01",
    freq="MS"
)

payroll_rows = []

payroll_id = 1

for month in months:

    for _, emp in employees.iterrows():

        min_sal, max_sal = salary_lookup[
            emp["comp_band_id"]
        ]

        annual_salary = np.random.randint(
            min_sal,
            max_sal + 1
        )

        monthly_salary = round(
            annual_salary / 12,
            2
        )

        bonus = round(
            monthly_salary *
            np.random.uniform(0, 0.20),
            2
        )

        overtime = round(
            np.random.uniform(0, 5000),
            2
        )

        deductions = round(
            monthly_salary *
            np.random.uniform(0.05, 0.15),
            2
        )

        gross_pay = round(
            monthly_salary +
            bonus +
            overtime,
            2
        )

        net_pay = round(
            gross_pay - deductions,
            2
        )

        payroll_rows.append({

            "payroll_fact_id": payroll_id,

            "employee_id": emp["employee_id"],

            "pay_period": month.strftime("%Y-%m"),

            "base_salary": monthly_salary,

            "bonus_amount": bonus,

            "overtime_amount": overtime,

            "deductions": deductions,

            "gross_pay": gross_pay,

            "net_pay": net_pay,

            "payment_status": np.random.choice(
                ["Paid","Paid","Paid","Paid","Delayed"],
                p=[0.24,0.24,0.24,0.24,0.04]
            )
        })

        payroll_id += 1

payroll = pd.DataFrame(payroll_rows)

print(payroll.shape)
```

```python
print(payroll.head())

print("Rows:", len(payroll))

print(
    "Unique Employees:",
    payroll["employee_id"].nunique()
)
```

```python
payroll.to_csv(
    "../datasets/facts/Fact_Payroll.csv",
    index=False
)

print("Fact_Payroll Saved Successfully")
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

months = pd.date_range(
    start="2023-01-01",
    end="2025-12-01",
    freq="MS"
)

attendance_rows = []

attendance_id = 1

for month in months:

    for _, emp in employees.iterrows():

        working_days = np.random.randint(20, 24)

        absent_days = np.random.randint(0, 4)

        leave_days = np.random.randint(0, 3)

        present_days = (
            working_days
            - absent_days
            - leave_days
        )

        present_days = max(present_days, 0)

        late_days = np.random.randint(0, 6)

        overtime_hours = round(
            np.random.uniform(0, 30),
            2
        )

        attendance_pct = round(
            (present_days / working_days) * 100,
            2
        )

        attendance_rows.append({

            "attendance_fact_id": attendance_id,

            "employee_id": emp["employee_id"],

            "attendance_month":
                month.strftime("%Y-%m"),

            "working_days": working_days,

            "present_days": present_days,

            "absent_days": absent_days,

            "leave_days": leave_days,

            "late_days": late_days,

            "overtime_hours": overtime_hours,

            "attendance_percentage":
                attendance_pct
        })

        attendance_id += 1

attendance = pd.DataFrame(
    attendance_rows
)

print(attendance.shape)
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

review_dates = pd.date_range(
    start="2023-01-01",
    end="2025-12-31",
    freq="MS"
)

performance_rows = []

for perf_id in range(1, 20001):

    employee_id = np.random.randint(
        1,
        5001
    )

    performance_score = round(
        np.random.normal(3.5, 0.8),
        2
    )

    performance_score = min(
        max(performance_score, 1),
        5
    )

    goal_achievement = round(
        np.random.uniform(60, 120),
        2
    )

    competency_score = round(
        np.random.uniform(1, 5),
        2
    )

    manager_rating = round(
        np.random.uniform(1, 5),
        2
    )

    if performance_score >= 4.5:
        category = "Outstanding"
    elif performance_score >= 4.0:
        category = "Exceeds Expectations"
    elif performance_score >= 3.0:
        category = "Meets Expectations"
    elif performance_score >= 2.0:
        category = "Needs Improvement"
    else:
        category = "Poor"

    performance_rows.append({

        "performance_fact_id": perf_id,

        "employee_id": employee_id,

        "review_date":
            np.random.choice(review_dates),

        "review_period":
            np.random.choice(
                ["Q1","Q2","Q3","Q4"]
            ),

        "performance_score":
            performance_score,

        "goal_achievement_pct":
            goal_achievement,

        "competency_score":
            competency_score,

        "manager_rating":
            manager_rating,

        "performance_category":
            category
    })

performance = pd.DataFrame(
    performance_rows
)

print(performance.shape)
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

change_dates = pd.date_range(
    start="2023-01-01",
    end="2025-12-31",
    freq="D"
)

change_types = [
    "Annual Increment",
    "Promotion",
    "Market Adjustment",
    "Role Change"
]

reasons = [
    "Performance Based",
    "Retention Strategy",
    "Market Correction",
    "Promotion Cycle",
    "Internal Mobility"
]

comp_changes = []

for i in range(1, 8001):

    emp_id = np.random.randint(1, 5001)

    old_salary = np.random.randint(30000, 200000)

    change_type = np.random.choice(change_types, p=[0.5, 0.2, 0.2, 0.1])

    if change_type == "Promotion":
        change_percent = np.random.uniform(10, 40)
    elif change_type == "Annual Increment":
        change_percent = np.random.uniform(3, 12)
    else:
        change_percent = np.random.uniform(-5, 15)

    new_salary = round(old_salary * (1 + change_percent/100), 2)

    comp_changes.append({

        "comp_change_fact_id": i,

        "employee_id": emp_id,

        "change_date": np.random.choice(change_dates),

        "change_type": change_type,

        "old_salary": old_salary,

        "new_salary": new_salary,

        "change_amount": round(new_salary - old_salary, 2),

        "change_percent": round(change_percent, 2),

        "reason": np.random.choice(reasons),

        "approved_by_manager_id": np.random.randint(1, 201)
    })

comp_change = pd.DataFrame(comp_changes)

print(comp_change.shape)
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

months = pd.date_range(
    start="2023-01-01",
    end="2025-12-01",
    freq="MS"
)

productivity_rows = []

prod_id = 1

for month in months:

    for _, emp in employees.iterrows():

        output_score = round(np.random.uniform(50, 100), 2)

        tasks_completed = np.random.randint(5, 50)

        quality_score = round(np.random.uniform(1, 5), 2)

        efficiency_score = round(np.random.uniform(1, 5), 2)

        kpi = round(np.random.uniform(60, 120), 2)

        if kpi >= 110:
            rating = "Excellent"
        elif kpi >= 90:
            rating = "Good"
        elif kpi >= 70:
            rating = "Average"
        else:
            rating = "Poor"

        productivity_rows.append({

            "productivity_fact_id": prod_id,

            "employee_id": emp["employee_id"],

            "productivity_month": month.strftime("%Y-%m"),

            "output_score": output_score,

            "tasks_completed": tasks_completed,

            "quality_score": quality_score,

            "efficiency_score": efficiency_score,

            "kpi_achievement_pct": kpi,

            "productivity_rating": rating
        })

        prod_id += 1

productivity = pd.DataFrame(productivity_rows)

print(productivity.shape)
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

exit_dates = pd.date_range(
    start="2023-01-01",
    end="2025-12-31",
    freq="D"
)

exit_types = [
    "Resignation",
    "Termination",
    "Retirement",
    "Layoff"
]

exit_reasons = [
    "Better Opportunity",
    "Compensation Issue",
    "Work-Life Balance",
    "Manager Conflict",
    "Relocation",
    "Career Growth",
    "Performance Issue"
]

exit_rows = []

for i in range(1, 1001):

    emp_id = np.random.randint(1, 5001)

    hire_date = pd.Timestamp("2023-01-01") + pd.to_timedelta(
        np.random.randint(0, 1096),
        unit="D"
    )

    exit_date = np.random.choice(exit_dates)

    tenure_days = max((pd.Timestamp(exit_date) - hire_date).days, 30)

    exit_rows.append({

        "exit_fact_id": i,

        "employee_id": emp_id,

        "exit_date": exit_date,

        "exit_type": np.random.choice(exit_types, p=[0.75, 0.15, 0.05, 0.05]),

        "exit_reason": np.random.choice(exit_reasons),

        "tenure_days": tenure_days,

        "exit_interview_score": round(np.random.uniform(1, 5), 2),

        "re_hire_eligibility": np.random.choice(["Yes", "No"], p=[0.3, 0.7]),

        "replacement_hired_flag": np.random.choice([0, 1], p=[0.4, 0.6])
    })

exit_df = pd.DataFrame(exit_rows)

print(exit_df.shape)
```
